```python
import numpy as np

x = np.array([-7,-4,-2,0,2])
y = np.array([1,2,2,0,-1])
```

```python
X = []
X.append(np.ones_like(x))
X.append(x)
```

```python
X = np.array(X).T
```

```python
X # a_0*1+a_1*x
```

```python
# X^T*X a = X^T*y

XTX = X.T@X
XTy = X.T@y

a = np.linalg.solve(XTX,XTy)

XTX
```

```python
a
```

```python
x_g = np.linspace(-7,2,100)
y_pred = a[0]+a[1]*x_g

import matplotlib.pyplot as plt

plt.plot(x_g,y_pred)
plt.scatter(x,y,c = 'r')
```
